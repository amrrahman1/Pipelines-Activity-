```python

import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.compose import make_column_selector, make_column_transformer
from sklearn.linear_model import LinearRegression
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.pipeline import make_pipeline
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python

path= '/content/drive/MyDrive/Sherlin machine learning/Cereal with missing values.xlsx'
df= pd.read_excel(path)
df.head()
```

```python
df = df.filter(['Manufacturer','type','calories per serving', 'grams of fat','grams of sugars','Weight in ounces per one serving'], axis=1)
```

```python
df.head()
```

```python
X = df.drop(columns=['calories per serving'])
y = df['calories per serving']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
cat_selector = make_column_selector(dtype_include='object')
num_selector = make_column_selector(dtype_include='number')
```

```python
mean_imputer = SimpleImputer(strategy='mean')
scaler = StandardScaler()
freq_imputer = SimpleImputer(strategy='most_frequent')
ohe= OneHotEncoder(sparse=False, handle_unknown='ignore')
```

```python
num_pipe = make_pipeline(mean_imputer, scaler)
cat_pipe = make_pipeline(freq_imputer,ohe)
```

```python
num_tuple = (num_pipe, num_selector)
cat_tuple = (cat_pipe, cat_selector)
```

```python
preprocess_data = make_column_transformer(num_tuple,cat_tuple)
```

```python
preprocess_data.fit(X_train)
```

```python
X_train_processed = preprocess_data.transform(X_train)
X_test_processed = preprocess_data.transform(X_test)
```

```python
print(np.isnan(X_train_processed).sum().sum(), 'missing values in training data')
print(np.isnan(X_test_processed).sum().sum(), 'missing values in testing data')
print('\n')
print('All data in X_train_processed are', X_train_processed.dtype)
print('All data in X_test_processed are', X_test_processed.dtype)
print('\n')
print('shape of data is', X_train_processed.shape)
print('\n')
X_train_processed
```
